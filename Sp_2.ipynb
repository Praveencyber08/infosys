```python
# ========== Real-Data Dashboard – Single Cell ==========
# Replace the API key placeholders below with your real keys BEFORE running.
# - NEWSAPI_KEY: NewsAPI.org key (for news)
# - TWITTER_BEARER_TOKEN: Twitter/X API v2 bearer token
# - OPENAI_API_KEY: OpenAI API key (for GPT sentiment)
# Optional: install pytrends if you want real Google Trends

# --------- Imports ---------
import os, time, json, textwrap
from datetime import datetime, timedelta
import requests
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from textblob import TextBlob
import yfinance as yf
from PIL import Image

# LLM and Twitter libraries
try:
    import openai
except Exception:
    openai = None
try:
    import tweepy
except Exception:
    tweepy = None

# Optional pytrends (real Google Trends)
try:
    from pytrends.request import TrendReq
    pytrends_available = True
except Exception:
    pytrends_available = False

# --------- CONFIG – PUT YOUR KEYS HERE ----------
NEWSAPI_KEY = "YOUR_NEWSAPI_KEY"            # <-- paste your NewsAPI key
TWITTER_BEARER_TOKEN = "YOUR_TWITTER_BEARER_TOKEN"  # <-- paste your Twitter/X Bearer token
OPENAI_API_KEY = "YOUR_OPENAI_API_KEY"      # <-- paste your OpenAI key (for GPT sentiment)

# Example local reference image path (you uploaded this earlier)
EXAMPLE_IMAGE_PATH = "/mnt/data/734655bf-327e-4fd5-bcbc-054bb619f15d.png"

# Companies list (extendable)
COMPANIES = [
    "Apple", "Google", "Amazon", "Microsoft", "Meta",
    "Tesla", "Samsung", "NVIDIA", "IBM", "Intel"
]

# Sector mapping (demo)
SECTORS = {
    "Apple":"Technology","Google":"Technology","Amazon":"Retail","Microsoft":"Technology",
    "Meta":"Social","Tesla":"Automobile","Samsung":"Technology","NVIDIA":"Semiconductor",
    "IBM":"Technology","Intel":"Semiconductor"
}

# Stock symbol map (yfinance)
SYMBOLS = {
    "Apple":"AAPL","Google":"GOOGL","Amazon":"AMZN","Microsoft":"MSFT",
    "Meta":"META","Tesla":"TSLA","Samsung":"005930.KS","NVIDIA":"NVDA",
    "IBM":"IBM","Intel":"INTC"
}

# Configure OpenAI
if OPENAI_API_KEY and openai is not None:
    openai.api_key = OPENAI_API_KEY

# --------- Helper Functions ---------
def fetch_news_newsapi(query, from_days=3, page_size=20):
    """Fetch recent news via NewsAPI (returns list of dicts)."""
    if not NEWSAPI_KEY:
        return []
    url = "https://newsapi.org/v2/everything"
    params = {
        "q": query,
        "from": (datetime.utcnow() - timedelta(days=from_days)).isoformat(),
        "language": "en",
        "sortBy": "publishedAt",
        "pageSize": page_size,
        "apiKey": NEWSAPI_KEY
    }
    try:
        r = requests.get(url, params=params, timeout=10)
        r.raise_for_status()
        return r.json().get("articles", [])
    except Exception as e:
        print("NewsAPI fetch error for", query, e)
        return []

def fetch_tweets_tweepy(query, max_results=50):
    """Fetch tweets with Tweepy (Twitter v2)."""
    if not TWITTER_BEARER_TOKEN or tweepy is None:
        return []
    try:
        client = tweepy.Client(bearer_token=TWITTER_BEARER_TOKEN, wait_on_rate_limit=True)
        # Basic recent search - adjust query to filter languages/retweets if needed
        query_str = f"{query} lang:en -is:retweet"
        resp = client.search_recent_tweets(query=query_str, max_results=min(100, max_results),
                                          tweet_fields=["created_at","lang","public_metrics","id","text"])
        if not resp or not resp.data:
            return []
        out = []
        for t in resp.data:
            out.append({"id": t.id, "text": t.text, "created_at": t.created_at.isoformat() if t.created_at else None})
        return out
    except Exception as e:
        print("Tweepy error:", e)
        return []

def fetch_tweets_snscrape(query, max_tweets=50):
    """Fallback: use snscrape if tweepy unavailable (no keys)."""
    try:
        import snscrape.modules.twitter as sntwitter
    except Exception:
        return []
    out = []
    for i, tweet in enumerate(sntwitter.TwitterSearchScraper(query).get_items()):
        if i >= max_tweets:
            break
        out.append({"id": tweet.id, "text": tweet.content, "created_at": tweet.date.isoformat()})
    return out

def openai_sentiment_json(text, model="gpt-4o-mini"):
    """Call OpenAI to get structured sentiment (label, score, reason)."""
    if not OPENAI_API_KEY or openai is None:
        return None
    prompt = f"""
Analyze the sentiment of the following text. Return ONLY a JSON object with keys:
'label' (Positive/Neutral/Negative),
'score' (number between -1.0 and 1.0),
'reason' (one-sentence explanation).
Text: \"\"\"{text}\"\"\"
"""
    try:
        # Use chat completion
        resp = openai.ChatCompletion.create(
            model=model,
            messages=[{"role":"user","content":prompt}],
            max_tokens=80,
            temperature=0.0
        )
        out = resp.choices[0].message.content.strip()
        # Try parse JSON substring
        try:
            j = json.loads(out)
            return j
        except Exception:
            # Try to find a JSON block inside output
            import re
            m = re.search(r"\{.*\}", out, re.DOTALL)
            if m:
                try:
                    return json.loads(m.group(0))
                except:
                    pass
        # Fallback to TextBlob if parse fails
        tb = TextBlob(text).sentiment.polarity
        lab = "Positive" if tb>0.1 else "Negative" if tb<-0.1 else "Neutral"
        return {"label": lab, "score": float(round(tb,3)), "reason": "fallback-textblob"}
    except Exception as e:
        print("OpenAI call error:", e)
        return None

def hf_or_textblob_sentiment(text):
    """Fallback sentiment using transformers if available, else TextBlob."""
    try:
        from transformers import pipeline
        hf = pipeline("sentiment-analysis")
        out = hf(text[:512])[0]
        label = out["label"]
        score = float(out.get("score", 0.0))
        if label.lower().startswith("positive"):
            sc = score
        elif label.lower().startswith("negative"):
            sc = -score
        else:
            sc = 0.0
        lab = "Positive" if sc>0.05 else "Negative" if sc<-0.05 else "Neutral"
        return {"label": lab, "score": round(sc,3), "reason": "hf-model"}
    except Exception:
        p = TextBlob(text).sentiment.polarity
        lab = "Positive" if p>0.1 else "Negative" if p<-0.1 else "Neutral"
        return {"label": lab, "score": float(round(p,3)), "reason": "textblob"}

def fetch_google_trends(keywords, timeframe="today 3-m"):
    """Try real pytrends if available, else simulate trends."""
    if pytrends_available:
        try:
            pytrends = TrendReq(hl='en-US', tz=360)
            pytrends.build_payload(keywords, timeframe=timeframe)
            data = pytrends.interest_over_time()
            if data.empty:
                raise Exception("empty pytrends")
            return data.drop(columns=['isPartial'], errors='ignore')
        except Exception as e:
            print("pytrends error:", e)
    # Simulate trends
    dates = pd.date_range(end=datetime.today(), periods=90)
    sim = pd.DataFrame(index=dates)
    for kw in keywords:
        v = np.abs(np.random.normal(loc=40, scale=12, size=len(dates)))
        # peaks
        for _ in range(np.random.randint(0,3)):
            center = np.random.randint(5, len(dates)-5)
            v += np.exp(-0.5*((np.arange(len(dates))-center)/3.0)**2) * np.random.randint(10,80)
        sim[kw] = (v / v.max() * 100).round(1)
    return sim

def fetch_stock_close(symbol, period_days=120):
    """Get recent close prices via yfinance (no key needed)."""
    try:
        df = yf.download(symbol, period=f"{period_days}d", progress=False)
        if df.empty:
            return pd.DataFrame()
        df = df.reset_index()[["Date","Close"]].rename(columns={"Date":"date","Close":"close"})
        df["date"] = pd.to_datetime(df["date"])
        return df.sort_values("date")
    except Exception as e:
        print("yfinance error for", symbol, e)
        return pd.DataFrame()

# --------- Data Collection & Aggregation ----------
print("Starting collection: news, tweets, sentiment, stocks, trends ...")
rows = []

# 1) Fetch news articles per company
for comp in COMPANIES:
    print("Fetching news for", comp)
    arts = fetch_news_newsapi(comp, from_days=3, page_size=10)
    if not arts:
        # create mock articles if NewsAPI missing or returned nothing
        arts = [{"title": f"{comp} posts quarterly results beating expectations", "publishedAt": datetime.utcnow().isoformat()},
                {"title": f"{comp} expands into new region", "publishedAt": (datetime.utcnow()-timedelta(hours=8)).isoformat()}]
    for a in arts:
        text = a.get("title") or a.get("description") or ""
        timestamp = a.get("publishedAt", datetime.utcnow().isoformat())
        sentiment = None
        if OPENAI_API_KEY and openai is not None:
            sentiment = openai_sentiment_json(text)
        if sentiment is None:
            sentiment = hf_or_textblob_sentiment(text)
        rows.append({
            "timestamp": timestamp,
            "source":"news",
            "company": comp,
            "text": text,
            "sent_label": sentiment["label"],
            "sent_score": sentiment["score"],
            "sent_reason": sentiment.get("reason")
        })

# 2) Fetch tweets per company (prefer Tweepy, fallback snscrape, otherwise mock)
for comp in COMPANIES:
    print("Fetching tweets for", comp)
    tweets = []
    if TWITTER_BEARER_TOKEN and tweepy is not None:
        tweets = fetch_tweets_tweepy(comp, max_results=80)
    if not tweets:
        tweets = fetch_tweets_snscrape(comp + " lang:en", max_tweets=40) if 'fetch_tweets_snscrape' in globals() else []
    if not tweets:
        tweets = [{"text": f"{comp} stock surges after product news", "created_at": datetime.utcnow().isoformat()} for _ in range(4)]
    for t in tweets[:40]:
        text = t.get("text","")
        timestamp = t.get("created_at", datetime.utcnow().isoformat())
        sentiment = None
        if OPENAI_API_KEY and openai is not None:
            sentiment = openai_sentiment_json(text)
        if sentiment is None:
            sentiment = hf_or_textblob_sentiment(text)
        rows.append({
            "timestamp": timestamp,
            "source":"twitter",
            "company": comp,
            "text": text,
            "sent_label": sentiment["label"],
            "sent_score": sentiment["score"],
            "sent_reason": sentiment.get("reason")
        })

# 3) Build DataFrame
df = pd.DataFrame(rows)
df["timestamp"] = pd.to_datetime(df["timestamp"], errors="coerce")
df = df.sort_values("timestamp", ascending=False).reset_index(drop=True)

# 4) Trends (pytrends or simulated)
print("Fetching Google Trends (pytrends if available, else simulated)...")
trend_df = fetch_google_trends(COMPANIES, timeframe="today 3-m")
# If trend_df is wide (columns per company) keep as-is, otherwise adapt

# 5) Stock closes (overview)
print("Fetching stock price overviews (yfinance)...")
stock_close = {c: fetch_stock_close(SYMBOLS.get(c,"AAPL"), period_days=90) for c in COMPANIES}

# 6) Analyses summary: current interest & sentiment aggregates
# current interest (use most recent trend point if available)
if isinstance(trend_df, pd.DataFrame) and not trend_df.empty:
    # if pytrends output has columns matching companies
    latest = trend_df.iloc[-1]
    try:
        current_interest = {c:int(latest[c]) for c in COMPANIES}
    except Exception:
        # columns may be per-keyword; fallback to simulated
        current_interest = {c: int(np.random.randint(5,30)) for c in COMPANIES}
else:
    current_interest = {c: int(np.random.randint(5,30)) for c in COMPANIES}

current_interest_df = pd.DataFrame({"Company": list(current_interest.keys()), "Interest": list(current_interest.values())})
current_interest_df = current_interest_df.sort_values("Interest", ascending=False)

# Sentiment counts per company
sent_counts = df.groupby(["company","sent_label"]).size().unstack(fill_value=0)
# overall sector distribution
sector_counts = pd.Series([SECTORS.get(c,"Other") for c in COMPANIES]).value_counts()

# 7) Save aggregated data
os.makedirs("data", exist_ok=True)
outpath = os.path.join("data", f"market_aggregate_{datetime.utcnow().strftime('%Y%m%d_%H%M%S')}.csv")
df.to_csv(outpath, index=False)
print("Saved aggregated data to:", outpath)

# --------- Build Matplotlib Dashboard (Option C style) ----------
sns.set_style("whitegrid")
plt.rcParams.update({'figure.max_open_warning': 0})

fig = plt.figure(figsize=(18,10))
fig.suptitle("Real-Time Industry Insight & Strategic Intelligence System", fontsize=16, fontweight='bold')

# Top: Trends (full width)
ax_trend = plt.subplot2grid((3,6),(0,0), colspan=6)
if isinstance(trend_df, pd.DataFrame) and not trend_df.empty:
    # if trend_df columns are company names
    for c in COMPANIES:
        if c in trend_df.columns:
            ax_trend.plot(trend_df.index, trend_df[c], label=c, linewidth=1.6, alpha=0.9)
        else:
            # if simulated, try to find matching column names ignoring case
            for col in trend_df.columns:
                if col.lower().startswith(c.lower()[:4]):
                    ax_trend.plot(trend_df.index, trend_df[col], label=c, linewidth=1.6, alpha=0.9)
                    break
else:
    # fallback simulated trend lines
    dates = pd.date_range(end=datetime.today(), periods=90)
    for c in COMPANIES:
        v = np.abs(np.random.normal(loc=40, scale=12, size=len(dates)))
        ax_trend.plot(dates, v/v.max()*100, label=c, linewidth=1.2, alpha=0.9)
ax_trend.set_title("Google Trends: Interest Over Time (Past 3 Months)")
ax_trend.set_ylabel("Search Interest")
ax_trend.legend(loc="upper right", fontsize=8, ncol=2)
ax_trend.tick_params(axis='x', rotation=25)

# Bottom-left: Current Search Interest
ax_a = plt.subplot2grid((3,6),(1,0), colspan=2, rowspan=2)
sns.barplot(x="Company", y="Interest", data=current_interest_df, ax=ax_a, palette="tab10")
ax_a.set_title("Current Search Interest")
ax_a.set_xticklabels(ax_a.get_xticklabels(), rotation=45, ha='right')

# Bottom-middle: News Sentiment (stacked)
ax_b = plt.subplot2grid((3,6),(1,2), colspan=2, rowspan=2)
for col in ["Positive","Neutral","Negative"]:
    if col not in sent_counts.columns:
        sent_counts[col] = 0
sent_counts_sorted = sent_counts.reindex(current_interest_df["Company"]).fillna(0)
companies_order = sent_counts_sorted.index.tolist()
p1 = ax_b.bar(companies_order, sent_counts_sorted["Positive"], label="Positive", color="#66c2a5")
p2 = ax_b.bar(companies_order, sent_counts_sorted["Neutral"], bottom=sent_counts_sorted["Positive"], label="Neutral", color="#8da0cb")
p3 = ax_b.bar(companies_order, sent_counts_sorted["Negative"], bottom=sent_counts_sorted.get("Positive",0)+sent_counts_sorted.get("Neutral",0), label="Negative", color="#fc8d62")
ax_b.set_title("News & Social Sentiment (stacked counts)")
ax_b.set_xticklabels(ax_b.get_xticklabels(), rotation=45, ha='right')
ax_b.set_ylabel("Count")
ax_b.legend()

# Bottom-right top: Sector distribution pie
ax_c = plt.subplot2grid((3,6),(1,4), colspan=1, rowspan=1)
ax_c.pie(sector_counts.values, labels=sector_counts.index, autopct="%1.1f%%", startangle=140)
ax_c.set_title("Sector Distribution")

# Bottom-right bottom: Recent News Highlights (text)
ax_d = plt.subplot2grid((3,6),(1,5), colspan=1, rowspan=2)
ax_d.axis("off")
ax_d.set_title("Recent News Highlights", pad=10)
# compile recent headlines (top 2 per company)
recent_headlines = []
for c in COMPANIES:
    sample = df[df["company"]==c]["text"].dropna().unique().tolist()[:2]
    for s in sample:
        recent_headlines.append(f"{c}: {s}")
# limit and format
recent_headlines = recent_headlines[:12]
text_str = "\n\n".join([textwrap.fill(f"{i+1}. {s}", width=28) for i,s in enumerate(recent_headlines)])
ax_d.text(0, 1, text_str, va="top", fontsize=9, family='monospace')

# Small stock overview inset
ax_stock = fig.add_axes([0.52, 0.04, 0.43, 0.18])
for c in COMPANIES:
    dfc = stock_close.get(c)
    if dfc is not None and not dfc.empty:
        ax_stock.plot(dfc["date"], dfc["close"], linewidth=1, label=c, alpha=0.8)
ax_stock.set_title("Stock Closing Prices (last ~90 days) - overview", fontsize=9)
ax_stock.tick_params(axis='x', labelrotation=20, labelsize=8)
ax_stock.tick_params(axis='y', labelsize=8)
ax_stock.legend(fontsize=7, ncol=3, loc='upper left')

plt.tight_layout(rect=[0, 0.03, 1, 0.95])

# show reference image small inset if exists
if os.path.exists(EXAMPLE_IMAGE_PATH):
    try:
        img = Image.open(EXAMPLE_IMAGE_PATH)
        ax_ref = fig.add_axes([0.02, 0.62, 0.18, 0.18])
        ax_ref.imshow(img)
        ax_ref.axis("off")
        ax_ref.set_title("Reference")
    except Exception as e:
        print("Could not load ref image:", e)

plt.show()

# Print short stats
print("\nRecords collected:", len(df))
print("Saved aggregated CSV at:", outpath)
```
